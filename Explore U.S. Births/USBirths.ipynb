```python
def read_csv(csv_file_name):
    data_string = open(csv_file_name).read()
    string_list = data_string.split('\n')[1:]
    final_list = []
    for item in string_list:
        int_fields = []
        string_fields = item.split(',')
        for elem in string_fields:
            int_fields.append(int(elem))
        final_list.append(int_fields)
    return final_list                         
```

```python
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
cdc_list[0:10]
```

```python
# a dictionary key is unique mobth and value is # of births
def month_births(input_list):
    Births_per_month = {}
    for item in input_list:
        Month = item[1]
        Births = item[4]
        if Month in Births_per_month:
            Births_per_month[Month] = Births_per_month[Month] + Births
        else:
            Births_per_month[Month] = Births
            
    return(Births_per_month)
cdc_month_births = month_births(cdc_list)
print(cdc_month_births)
```

```python
"""function that calculates total numnber of births for each unique 
day of week"""
def dow_births(input_list):
    Birth_per_day = {}
    for item in input_list:
        dow = item[3]
        bdow = item[4]
        if dow in Birth_per_day:
            Birth_per_day[dow] = Birth_per_day[dow] + bdow
        else:
            Birth_per_day[dow] = bdow
    return(Birth_per_day)
cdc_day_births = dow_births(cdc_list)
print(cdc_day_births)
```

```python
births = {}
def calc_counts(data, column):
    for item in data:
        col = item[column]
        bnum = item[4]
        if col in births:
            births[col] = births[col] + bnum
        else:
            births[col] = bnum
    return(births)

            
```

```python
cdc_year_births = calc_counts(cdc_list, 0)
cdc_month_births = calc_counts(cdc_list, 1)
cdc_dom_births = calc_counts(cdc_list, 2)
cdc_dow_births = calc_counts(cdc_list, 3)
```
